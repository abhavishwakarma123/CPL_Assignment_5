```python
#setting the function and variables

exp_1 = "math.log(x/2) - math.sin(5*x/2)"
a_1 = 1.6
b_1 = 2.7
eps_1 = 0.00001


#calculating the roots using bisection and regula falsi

root_1 = ml.Roots.bisection(exp_1, a_1, b_1, eps_1, True, True)
root_2 = ml.Roots.regula_falsi(exp_1, a_1, b_1, eps_1, True, True)


#printing the roots

print()
print()
print()
print("Root from bisection = ", round(root_1, 6))
print("Root from regula falsi = ", round(root_2, 6))

'''Clearly, regula falsi is much more efficient that bisection and takes less number of iterations'''
```

```python
#setting the expression and variables for calculation of root using bisection, regula falsi and newton raphson methods

exp_2 = "-x-math.cos(x)"

#since the root is when cos(x) = -x, hence we guess that the root lies between -1.57(pi/2) and 0

a_2 = -1.57                       #same interval for both bisection and regula falsi
b_2 = 0.0
eps_2 = 0.00001

#as asked, the first guess for newton raphson is taken to be zero
guess_1 = 0.0


#calculating the roots

root_3 = ml.Roots.bisection(exp_2, a_2, b_2, eps_2)
root_4 = ml.Roots.regula_falsi(exp_2, a_2, b_2, eps_2)
root_5 = ml.Roots.newton_raphson(exp_2, guess_1, eps_2)


#printing the roots

print("Root from bisection = ", round(root_3, 6))
print("Root from regula falsi = ", round(root_4, 6))
print("Root from newton raphson = ", round(root_5, 6))


#checking the roots

func_2 = lambda x: eval(exp_2)

print("Function at root from bisection = ", round(func_2(root_3), 6))
print("Function at root from regula falsi = ", round(func_2(root_4), 6))
print("Function at root from newton raphson = ", round(func_2(root_5), 6))

'''Clearly, the roots give function values close to zero'''
```

```python
#calculation of polynomial roots using laguerre method with synthetic divison for deflation


#setting the coefficient matrix of polynomial(highest to lowest order from left to right) and the guess value of 1st root
#given function is 1*x**3 - 5*x**3 + 4

coeff = [1, 0, -5, 0, 4]

#Looking at the equation, clearly x = 1 is a root, so we take it as first guess
guess_2 = 1


#writing the polynomial for printing and checking purpose
pol = ""
for i in range(len(coeff)-1):
    pol += f"({coeff[i]})*x**{len(coeff) - i - 1} + "
pol += f"{coeff[len(coeff) - 1]}"

print("The given equation is: ", pol)
print("\n")


#calculating the roots

root = ml.Roots.laguerre(coeff, guess_2)


#rounding off the roots to 6 decimal places

print("Roots of the given polynomial rounded to six decimal places are:")
for i in range(len(root)):
    print(round(root[i], 6), end = ' ')
print("\n")
    

#rounding off the roots to 2 decimal places

print("Roots of the given polynomial rounded to two decimal places are:")
for i in range(len(root)):
    print(round(root[i], 2), end = ' ')
print("\n")


#checking the roots

func_3 = lambda x: eval(pol)

print("Function values for the roots in the order given in the output:")

for i in range(len(root)):
    print(round(func_3(root[i]), 6), end = ' ')

'''Clearly, the roots give function values close to zero'''
```
